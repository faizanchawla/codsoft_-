# "To-Do List application"

```python
print("**********TO-DO LIST APPLICATION**************")

lists = []
task_id = 1
lists.append(task_id)
task = {}
description = {'task': '', 'description': ''}
task = dict.fromkeys(lists, description)


def add_task(task_description, task_detail):
    global task_id
    task[task_id] = {'task': task_description, 'description': task_detail}
    task_id += 1
    
def update_task(task_id, task_description, task_detail):
    if task_id in task:
        task[task_id] = {'task': task_description, 'description': task_detail}
    else:
        print("Task not Found")
    print("Updated task:", task)    

def list_task():
    for i,j in task.items():
        print(f"{i}: task: {j['task']}\n   detail {j['description']}")

while True:
    print("\n")
    print("1. Add a task")
    print("2. List tasks")
    print("3. Update a task")
    print("4. Quit")
    print("\n")

    choice = input("Enter your choice: ")

    if choice == "1":
        t = input("Enter task: ")
        d = input("Enter Task Description")
        add_task(t, d)
    elif choice == "2":
        list_task()
    elif choice == "3":
        task_id = int(input("Enter task ID to update: "))
        new_task = input("Enter task: ")
        new_description = input("Enter new Task Description: ")
        update_task(task_id, new_task, new_description,)
    elif choice == "4":
        break
    else:
        print("Invalid choice. Please try again.")

# # Adding a task
# add_task("code", "python")
# add_task("c", "p")
# # update_task(1, "c", "p")
# list_task()
```

# "CALCULATOR"

```python
print("*******************Simple Calculator***********************")

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Cannot divide by zero"
    else:
        return a / b
    
def power(a, b):
    return a**b

while True:
    try:
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))

        print("Select operation:")
        print("     1. Addition")
        print("     2. Subtraction")
        print("     3. Multiplication")
        print("     4. Division")
        print("     5. Power")
        print("\n")

        operation = int(input("Enter operation (1/2/3/4/5): "))

        if operation == 1:
            result = add(num1, num2)
            print(f"Result: {num1} + {num2} = ", result)
        elif operation == 2:
            result = subtract(num1, num2)
            print(f"Result: {num1} - {num2} = ", result)
        elif operation == 3:
            result = multiply(num1, num2)
            print(f"Result: {num1} * {num2} = ", result)
        elif operation == 4:
            result = divide(num1, num2)
            print(f"Result: {num1} / {num2} = ", result)
        elif operation == 5:
            result = power(num1, num2)
            print(f"Result: {num1} ** {num2} = ", result)

        else:
            print("Invalid input. Please select a valid operation.")

        cont = input("Do you want to perform another calculation? (y/n): ").lower()
        if cont != 'y':
            break
    except ValueError:
        print("Invalid input. Please enter a valid number.")
```

```python
import string
import random

def generate_password(length):
    characters = string.ascii_letters + string.digits
    password = ''.join(random.choice(characters) for i in range(length))
    return password

def main():
    print("Password Generator")

    while True:
        try:
            length = int(input("Enter the desired length of the password: "))

            if length < 1:
                print("Password length should be at least 1.")
                continue

            password = generate_password(length)
            print("Generated Password:", password)

            cont = input("Do you want to generate another password? (yes/no): ").lower()
            if cont != 'yes':
                break
        except ValueError:
            print("Invalid input. Please enter a valid length (a positive integer).")

if __name__ == "__main__":
    main()
```
